```python
import json
f = open('cheersyou_offer.json')
data = json.load(f)
```

```python
cheersyou_offer = pd.DataFrame(columns = ["name", "GPA", "GRE","IELTS","TOEFL", "Admission_school",
                                "Admission_major", "undergraduate_school", 
                                "undergraduate_major"])


names = ["cheersyou" + str(x) for x in list(range(1,283+1))]

undergraduate = []
for student in data['info']:
    if '毕业于' in student:
        undergraduate.append(student['毕业于'])
    else:
        undergraduate.append("Nan")
        
GPA = []
for student in data['info']:
    if 'GPA' in student:
        GPA.append(student['GPA'])
    else:
        GPA.append("Nan")

GRE = []
for student in data['info']:
    if 'GRE' in student:
        GRE.append(student['GRE'])
    else:
        GRE.append("Nan")

GRE = []
for student in data['info']:
    if 'GRE' in student:
        GRE.append(student['GRE'])
    else:
        GRE.append("Nan")

GRE = []
for student in data['info']:
    if 'GRE' in student:
        GRE.append(student['GRE'])
    else:
        GRE.append("Nan")
        
Admission_school = []
for student in data['info']:
    if 'offer' in student:
        Admission_school.append(student['offer'])
    else:
        Admission_school.append("Nan")
        
cheersyou_offer["name"] = names
cheersyou_offer["undergraduate_school"] = undergraduate
cheersyou_offer["GPA"] = GPA
cheersyou_offer["GRE"] = GRE
cheersyou_offer["Admission_school"] = Admission_school

cheersyou_offer = cheersyou_offer.explode('Admission_school').reset_index(drop=True)

cheersyou_offer["Admission_major"] = cheersyou_offer['Admission_school'].apply(lambda x: re.sub(r'[^a-zA-Z]', '', str(x)))

for i in range(0,len(cheersyou_offer['Admission_school'])):
    cheersyou_offer['Admission_school'][i] = re.search("(\p{Han}*)", str(cheersyou_offer['Admission_school'][i])).group(1)
```

```python
f = open('palmdrive_offer.json')
data = json.load(f)
len(data["offer_info"])
```

```python
palmdrive_offer = pd.DataFrame(columns = ["name", "GPA", "GRE","IELTS","TOEFL", "Admission_school",
                                "Admission_major", "undergraduate_school", 
                                "undergraduate_major"])

names = ["palmdrive" + str(x) for x in list(range(1,32+1))]

admission_school = []
for student in data['offer_info']:
    if 'admission_school' in student:
        admission_school.append(student['admission_school'])
    else:
        admission_school.append("Nan")
        
admission_major = []
for student in data['offer_info']:
    if 'admission_major' in student:
        admission_major.append(student['admission_major'])
    else:
        admission_major.append("Nan")

undergraduate = []
for student in data['offer_info']:
    if 'undergraduate' in student:
        undergraduate.append(student['undergraduate'])
    else:
        undergraduate.append("Nan")

GPA = []
GRE = []
IELTS = []
TOEFL = []
for student in data['offer_info']:
    if 'background' in student:
        GPA.append(re.search("GPA：(.*?)，", student["background"]).group(1))
        GRE.append(re.search("GPA：(.*?)，", student["background"]).group(1) if re.search("GRE：(.*?)，", student["background"]) else "Nan")
        TOEFL.append(re.search("TOEFL：(.*?)，", student["background"]).group(1) if re.search("TOEFL：(.*?)，", student["background"]) else "Nan")
        IELTS.append(re.search("IELTS:?：?(.*?)，", student["background"]).group(1) if re.search("IELTS:?：?(.*?)，", student["background"]) else "Nan")

palmdrive_offer["name"] = names
palmdrive_offer["GPA"] = GPA
palmdrive_offer["GRE"] = GRE
palmdrive_offer["IELTS"] = IELTS
palmdrive_offer["TOEFL"] = TOEFL
palmdrive_offer["Admission_school"] = admission_school
palmdrive_offer["Admission_major"] = admission_major
palmdrive_offer["undergraduate_school"] = undergraduate
palmdrive_offer["GPA"] = palmdrive_offer["GPA"].apply(lambda x: x[0:3])
```

```python
meishiedu_offer = pd.DataFrame(columns = ["name", "GPA", "GRE","IELTS","TOEFL", "Admission_school",
                                "Admission_major", "undergraduate_school", 
                                "undergraduate_major"])

names = ["meishiedu" + str(x) for x in list(range(1,261+1))]
undergraduate = []
undergraduate_major = []
GPA = []
GRE = []
IELTS = []
TOEFL = []
Admission_school = []
Admission_major = []

for student in msd_offer["msd_offer_info"]:
    undergraduate.append(re.search("(.*)\/(.*)", student['背景院校/专业']).group(1) if re.search("(.*)\/(.*)", student['背景院校/专业']) else  student['背景院校/专业'])
    undergraduate_major.append(re.search("(.*)\/(.*)", student['背景院校/专业']).group(2) if re.search("(.*)\/(.*)", student['背景院校/专业']) else "Nan")
    Admission_school.append(student['录取院校/专业'])
    GPA.append(re.search("GPA.?([0-9]+\.?[0-9]+)", student['成绩说明']).group(1) if re.search("GPA.?([0-9]+\.?[0-9]+)", student['成绩说明']) else "Nan")
    GRE.append(re.search("GRE.?([0-9]+\+?[0-9]+)", student['成绩说明']).group(1) if re.search("GRE.?([0-9]+\+?[0-9]+)", student['成绩说明']) else "Nan")
    TOEFL.append(re.search("TOEFL.?([0-9]+\+?[0-9]+)", student['成绩说明']).group(1) if re.search("TOEFL.?([0-9]+\+?[0-9]+)", student['成绩说明']) else "Nan")
    IELTS.append(re.search("IELTS.?([0-9]+.?[0-9]?)", student['成绩说明']).group(1) if re.search("IELTS.?([0-9]+.?[0-9]?)", student['成绩说明']) else "Nan")
    
        

meishiedu_offer["undergraduate_school"] = undergraduate
meishiedu_offer["undergraduate_major"] = undergraduate_major
meishiedu_offer["Admission_school"] = Admission_school
meishiedu_offer["GPA"] = GPA
meishiedu_offer["GRE"] = GRE
meishiedu_offer["IELTS"] = IELTS
meishiedu_offer["TOEFL"] = TOEFL
meishiedu_offer["name"] = names
```
